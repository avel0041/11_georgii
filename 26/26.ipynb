```python
a=open('26_1.txt')
a=a.readlines()
a = list(map(int, a))
p=set(a)
c=0
max_s=0
for i in range(0,len(a)-1):
    for w in range(i+1, len(a)):
        if (a[i] + a[w])%2==0:
            s = a[i]+a[w]
            if (s in p):
                c=c+1
                if s>max_s:
                    max_s=s
print(max_s, c, sep='\n')

```
