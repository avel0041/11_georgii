```python
a=open('26_1.txt')
a=a.readlines()
a = list(map(int, a))
p=set(a)
c=0
max_s=0
for i in range(0,len(a)-1):
    for w in range(i+1, len(a)):
        if (a[i] + a[w])%2==0:
            s = a[i]+a[w]
            if (s in p):
                c=c+1
                if s>max_s:
                    max_s=s
print(max_s, c, sep='\n')
```

```python
f = open('26_2.txt')
s=f.readlines()
s=s[1:]
s=set(map(int, s))
c=0
c_max = 0
q=0
for i in range(min(s), max(s)):
    if i in s:
        c=c+1
        if c>c_max:
            c_max = c
            q=i-c+1
    else:
        c=0
print(q, c_max)
```

```python
import math
f = open('26_2_2.txt')
a=f.readlines()[1:]
c=0
sum=0
i_max=0
a=list(map(int, a))
s = []
for w in a:
    if w<=50:
        sum=sum+w
    else:
        s.append(w)

n=len(s)//2
s.sort()
for i in s:
    i=int(i)
    c=c+1
    if c<=n:
        if c==n:
            i_max=i
        i=i-i/4
        sum=sum+i
        
    if c>n:
        sum=sum+i

print(math.ceil(sum), i_max)
    

```

```python
f = open('26_3.txt')
s = f.readlines()
s=s[1:]
q=[[0]*10001 for w in range(0, 10001)]
for t in s:
    t=t.split(' ')
    y=int(t[0])
    x=int(t[1])
    q[y][x] = 1 # стрельба (замена в массиве)

k=0
k_max=0
c_max=0
for i in q:
    k=k+1
    c=0
    for w in i:
        if w==1:
            c=c+1
            if c_max<c:
                c_max=c
                k_max=k
        else:
            c=0

print(c_max, k_max-1)
    
```

```python
f = open('26_4.txt')
q=f.readlines()
q=q[1:]
start = 1633305600
a = [0] * 60 * 60 * 24 * 7
finish = 1633305600 + 60 * 60 * 24 * 7
for i in q:
    i = i.split(' ')
    i = list(map(int,i))
    if i[0]<=start and (i[1]>finish or i[1] ==0):
        a[0] = a[0] + 1
    if i[0]<=start and (i[1]<=finish and i[1]>start):
        a[0]=a[0]+1 
        a[i[1]-start] -= 1
    if (i[0]>start and i[0]<=finish) and (i[1]>finish or i[1] ==0):
        a[i[0]-start]+= 1
    if i[0]>start and (i[1]<=finish and i[1]>start):
        a[i[0]-start] +=1
        a[i[1]-start] -= 1

k=0
k_m=0
max_l = 0
for l in a:
    k=k+l
    if k>k_m:
        k_m=k
        max_l = 0
    if k == k_m:
        max_l += 1

print(k_m, max_l)


```
