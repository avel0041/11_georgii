```python
from turtle import *
color('black', 'red')
speed(100)
left(90)
k = 20
begin_fill()
for i in range(1):
    rt(90)
    circle(5*k, -180)
    left(90)
    circle(5*k, -180)
    left(90)
    circle(5*k, -180)
    left(90)
    circle(5*k, -180)

end_fill()
canvas = getcanvas()
wk = 0
for x in range(-100*k, 100*k, k):
    for y in range(-100*k, 100*k, k):
        s = canvas.find_overlapping(x,y,x,y)
        if len(s) == 1 and s[0] == 5:
            wk=wk+1
print(wk)

done()
exit()
```
