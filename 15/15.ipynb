```python
p  = [i for i in range (20,51)]
q  = [i for i in range (30,66)]
t=[]
for a_min in range(1, 100):
    for a_max in range(a_min + 1, 100):
        a = [i for i in range(a_min, a_max)]
        flag =1
        for x in range(1, 100):
            if ((x in a) or ((not(x in p)) or(not(x in q)))) !=1:
                flag =0
                break
        if flag ==1:
            t.append(a_max-a_min)
    
print(min(t)-1)
           

```
