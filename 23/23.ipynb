```python
a = []
def f(x, i):
    if i <= 5:
        a.append(x)
    if i < 5:
        f(x+1, i+1)
        f(x*2, i+1)

f(1, 0)
a = set(a)
print(sorted(a))
```

```python
def f(x, y):
    if x>y:
        return 0
    if x==y:
        return 1
    else:
        return f(x+1, y) + f(x+2, y)
    
print(f(3,9) * f(9, 14) * f(16,20))
```

```python
def f(n, i):
    if n>i:
        return f(n-2, i)+f(n//2,i)
    if n==i:
        return 1
    if n<i:
        return 0
    
print(f(38,16)*f(16,2))
```

```python
def f(n, i):
    if n==26:
        return 0
    if n<i:
        return f(n+1, i)+f(2*n+1,i)
    if n==i:
        return 1
    if n>i:
        return 0

print( f(1, 27))
```
