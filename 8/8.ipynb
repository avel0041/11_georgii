Сколько различных вариантов шифра можно задать, если известно, что цифра 1 встречается ровно два раза, а каждая из других допустимых цифр может встречаться в шифре любое количество раз или не встречаться совсем?

```python
from itertools import product
words = list(product('1234', repeat=5))
a=[]

for i in words:
    if i.count('1') == 2:
        a.append(i)
print(len(a))
```

Михаил составляет 5-⁠буквенные коды. В кодах разрешается использовать только буквы А, Б, В, Г, Д, при этом код не может начинаться с гласной и не может содержать двух одинаковых букв подряд. Сколько различных кодов может составить Михаил?

```python
from itertools import product
words = list(product('АБВГД', repeat=5))
a=[]
for i in words:
    if i[0] != 'А':
        flag = False
        for j in range(0, len(i) - 1):
            if i[j] == i[j + 1]:
                flag = True
        if flag == False:
            a.append(i)
print(len(a))
        
```

Алиса составляет 6-⁠буквенные слова из букв М, А, Н, Г, У, С, Т. Каждая из букв может встречаться сколько угодно раз, причём первой буквой не может быть А, буква У должна встречаться не менее 1 раза. Также в записи должны быть ровно две буквы М.

Сколько различных слов может составить Алиса?

```python
from itertools import product
words = list(product('МАНГУСТ', repeat=6))
a=[]
for i in words:
    if i[0] != 'А' and i.count('М')==2 and i.count('У') >= 1:
        a.append(i)

print(len(a))
        
```

Определите количество восьмизначных 15-⁠ричных чисел, в записи которых ровно два нуля и не более четырёх цифр, для записи которых используются буквы.

```python
from itertools import product
words = list(product('01A', repeat=8))
k = 0
for i in words:
    if i.count('0')==2 and i.count('A') <= 4 and i[0] != '0' :
        k += 9**i.count('1') * 5**i.count('A')
print(k)

        
```

Найдите количество пятизначных восьмеричных чисел, в которых все цифры различны и никакие две четные или нечетные не стоят рядом.

```python
import itertools
a=list(itertools.permutations('01234567', 5))
c=[]

for i in a:
    if i[0]=='0':
        continue
    flag = True
    for w in range(0, len(i)-1):
        if int(i[w])%2 == int(i[w+1])%2:
            flag = False
    if flag == True:
        c.append(i)
print(len(c))

            
 
```

Алфавит = {А, Б, В, Т, О, М}
Сколько слов в которых содержится слово ТОМ?

```python
from itertools import *
 
s = 'АБВТОМ'
a = list(permutations(s, 6))
c=0
for i in a:
    b = ''.join(i) #перевод массива(кортежа) в строчку
    if 'ТОМ' in b:
        c=c+1

print(c/len(a))
```
