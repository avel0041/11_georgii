```python
a=open('17_1.txt')
w=a.readlines()
count=0
r=[]
for i in range(0, len(w)-1):
    if int(w[i])%3==0 or int(w[i+1])%3==0:
        count=count+1
        e=int(w[i])+int(w[i+1])
        r.append(e)
print(count, max(r))
        
```

```python
a=open('17_2.txt')
s=a.readlines()
p=[]
c=0
for i in range(0, len(s)-2):
    r = sorted([int(s[i]), int(s[i+1]), int(s[i+2])])
    if ((r[0] < r[1]+r[2]) and ((r[0])**2 < (r[1])**2+(r[2])**2))==1:
        c=c+1
        p.append(sum(r))

print(c, max(p))
        
        
```
