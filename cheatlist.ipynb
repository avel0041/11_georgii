### Циклы

```python
# Цикл while
# Пока выполняется условие:
#   делай это

a = 0
while a < 10: #Пока a меньше 10
    print(a) #Выводи a
    a = a + 1 #Прибавь к а 1

# Цикл for
# Для каждого значения из указанного множества:
#   делай это

for a in range(0, 101, 1): #Для каждого а в диапазоне от 0 до 100 (101 не включительно!) с шагом 1
    print(a) #выведи этот a

for a in range(100, -1, -1): #Для каждого а в диапазоне от 100 до 0 (-1 не включительно!) с шагом -1
    print(a) #выведи этот a

a = '12345abc' # а может быть строкой
a = [1,2,7,4,10] # a может быть массивом
a = {1, 2, 4} # a может быть множеством
for el in a: #Для каждого элемента из a
    print(el) #выведи этот элемент


# Работа с индексами
a = [1, 2, 1, 4, 6]
for i in range(len(a)): #Перебираем индексы от 0 до длины массива (не включительно)
    print(a[i]) #Выводим элемент с индексом i



a = [1, 2, 3, 4, 4, 5]
for el in a:
    if el == 4: # Если элемент равен 4
        continue # переходим к следующему шагу цикла
    print(el)
```

### Черепашка (6 задание ЕГЭ)

```python
from turtle import * #Подключили библиотеку
color('black', 'red') #Установили цвета для линий (черный) и заливки (красный)
speed(100) # Скорость движения
left(90) # Повернули на 90 градусов, чтобы соответсвовать условию задачи
k = 100 # Шаг черепашки (100 чтобы было видно рисунок, иначе будет маленький)
begin_fill() # Начинаем заливку в рамках нашей фигуры
for i in range(3): # Чертим треугольник (нужно четко определить, когда закончить чертить (то есть нельзя 2 раза проходить по одной и той же линии)) 
    fd(10*k) # Вперед на 10 * шаг
    rt(120) # Направо на 120 градусов
    
end_fill() # Заканчиваем заливку
canvas = getcanvas() # Получаем нашу картинку
wk = 0 #Счетчик
for x in range(-100*k, 100*k, k): # Проходимся по каждой целой точке нашей картинки
    for y in range(-100*k, 100*k, k): # Проходимся по каждой целой точке нашей картинки
        s = canvas.find_overlapping(x,y,x,y) #Проверяем есть ли там то, что мы нарисовали, либо просто пустая точка
        if len(s) == 1 and s[0] == 5: # Если там что-то есть (длина = 1) и это что-то красного цвета (s[0] == 5 - это проверка на красный цвет (цвет заливки))
            wk=wk+1 #Прибавили счетчик
print(wk)
done()
exit()

# Другие функции
circle(5*k, -180) # Рисует дугу (градусная мера 180) с радиусом 5. Центр окружности слева от черепашки на расстоянии 5. Рисуем по часовой стрелке (если градусная мера положительная, то против часовой)
lt(90) # Поворот налево на 90 градусов
```
