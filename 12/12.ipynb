```python
import itertools
a=itertools.product('12', repeat=14)
q=[]
for i in a:
    
    if (10==i.count('1')) and (4==i.count('2')):
        s = ''
        for j in i:
            s += str(j)
        while ('11' in s):
            if '112' in s:
                s=s.replace('112', '6', 1)
            else:
                s=s.replace('11', '3', 1)
        r=0
        for t in s:
            r=r+int(t)
        q.append(r)
print(max(q))


```
