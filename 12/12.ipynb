```python
import itertools
a=itertools.product('12', repeat=14)
q=[]
for i in a:
    
    if (10==i.count('1')) and (4==i.count('2')):
        s = ''
        for j in i:
            s += str(j)
        while ('11' in s):
            if '112' in s:
                s=s.replace('112', '6', 1)
            else:
                s=s.replace('11', '3', 1)
        r=0
        for t in s:
            r=r+int(t)
        q.append(r)
print(max(q))


```

```python
# 01321313213123213210
# 010 -> 1030 -> 121100
# 020 -> 2010 -> 21030 -> 2121100
# 030 -> 21100
# x1 - количесвто 1
# x2 - количесвто 2
# x3 - количесвто 3

# 196 (количество 1 в конечной строке) = 3*x1 + 3*x2 + 2*x3
# 75 = x1 + 2*x2 + x3
# x1+x2+x3 -> min
a=[]
for x in range(100):
    for y in range(100):
        for z in range(100):
            if x+2*y+z==75 and 3*x+3*y+z*2==196:
                a.append(x+y+z)

print(min(a)+2)
```
