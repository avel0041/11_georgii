```python
def kl(p):
    w=[]
    for t in p:
        d=0
        for t1 in p:
            dt=((t[0]-t1[0])**2+(t[1]-t1[1])**2)**0.5
            d+=dt
        w.append([d, t[0], t[1]])
    return min(w)
    
a=open('27_B_5.txt')
a=a.readlines()
a1 = []
a2 = []
a3 = []
for i in a:
    i=i.split()
    b = list(map(float, i))
    x = b[0]
    y = b[1]
    if x<-6 and y>0:
        a1.append([x,y])
    elif y>0 and x>-6:
        a2.append([x,y])
    else:
        a3.append([x,y])

cr1=kl(a1)
x1 = cr1[1]
y1 = cr1[2]

cr2=kl(a2)
x2 = cr2[1]
y2 = cr2[2]

cr3=kl(a3)
x3 = cr3[1]
y3 = cr3[2]

v = [a1, a2, a3]
sum_x = 0
sum_y = 0
for el in v:
    d, x, y = kl(el)
    sum_x+=x
    sum_y+=y
print(sum_x/3, sum_y/3)

#print((x1+x2+x3)*10000/3, (y1+y2+y3)*10000/3)

```
